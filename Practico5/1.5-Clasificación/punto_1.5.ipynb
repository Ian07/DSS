```python
# Imports needed for the script
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt  
%matplotlib inline
import matplotlib.gridspec as gridspec
import itertools
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import train_test_split 
from sklearn.svm import SVC 
from sklearn.metrics import classification_report, confusion_matrix  
from sklearn.metrics import log_loss, auc, roc_curve

#******************************************

from itertools import cycle
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
from scipy import interp
```

```python
datos = pd.read_csv("bill_authentication.csv", encoding="utf-8", delimiter=",")
```

```python
X = datos.drop(['Class'], axis=1)  
y = datos['Class']  
```

```python
xs = X.values
ys = y.values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y) 
```

```python
svclassifier_lineal = SVC(kernel='linear', gamma='scale', probability=True)  # Kernel lineal
svclassifier_lineal.fit(X_train, y_train) 
y_pred_lineal = svclassifier_lineal.predict(X_test)
probas_lineal= svclassifier_lineal.fit(X_train, y_train).predict_proba(X_test)
```

```python
svclassifier_poly = SVC(kernel='poly', degree=4, gamma='scale', probability=True)  # Kernel polinómico
svclassifier_poly.fit(X_train, y_train) 
y_pred_poly = svclassifier_poly.predict(X_test)
probas_poly= svclassifier_poly.fit(X_train, y_train).predict_proba(X_test)
```

```python
svclassifier_rbf = SVC(kernel='rbf', gamma='scale', probability=True) # Kernel Gaussiano
svclassifier_rbf.fit(X_train, y_train) 
y_pred_rbf = svclassifier_rbf.predict(X_test)
probas_rbf= svclassifier_rbf.fit(X_train, y_train).predict_proba(X_test)
```

```python
# Compute ROC curve and area the curve
fpr, tpr, thresholds = roc_curve(y_test, probas_lineal[:, 1])
roc_auc = auc(fpr, tpr)
print ("Area under the ROC curve : %f" % roc_auc)
```

```python
# Plot  curva ROC kernel lineal
plt.clf()
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Curva ROC Lineal')
plt.legend(loc="lower right")
plt.show()
```

```python
# Compute ROC curve and area the curve
fpr, tpr, thresholds = roc_curve(y_test, probas_poly[:, 1])
roc_auc = auc(fpr, tpr)
print ("Area under the ROC curve : %f" % roc_auc)
```

```python
# Plot  curva ROC kernel Polinomial
plt.clf()
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Curva ROC Polinomial')
plt.legend(loc="lower right")
plt.show()
```

```python
# Compute ROC curve and area the curve
fpr, tpr, thresholds = roc_curve(y_test, probas_rbf[:, 1])
roc_auc = auc(fpr, tpr)
print ("Area under the ROC curve : %f" % roc_auc)
```

```python
# Plot  curva ROC kernel RBF
plt.clf()
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Curva ROC rbf')
plt.legend(loc="lower right")
plt.show()
```

```python
xs = xs[:,[0, 2]]

# Plotting Decision Regions
gs = gridspec.GridSpec(2, 2)
fig = plt.figure(figsize=(10, 8))

for clf, lab, grd in zip([svclassifier_poly, svclassifier_rbf, svclassifier_lineal],
                         ['Polinómico', 'RBF', 'Lineal'],
                         itertools.product([0, 1], repeat=2)):
    clf.fit(xs, ys)
    ax = plt.subplot(gs[grd[0], grd[1]])
    fig = plot_decision_regions(X=xs, y=ys, clf=clf, legend=2)
    plt.title(lab)
plt.show()
```
